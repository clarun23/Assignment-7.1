```python
## Assignment - Session 7 - problem 1
import pandas as pd
import numpy as np

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays':  [[23, 47], [], [24, 43, 87], [13], [67, 32]], 
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', '12. Air France', '"Swiss Air"']})

df
```

```python
## fill the missing flight numbers
df['FlightNumber'] = df['FlightNumber'].interpolate().astype(int)
df
```

```python
## split from and to
df['From'] = df['From_To'].str.split('_',1).str[0]
df['To'] = df['From_To'].str.split('_',1).str[1]
df1 = pd.DataFrame(df, columns = ['Airline', 'FlightNumber','From', 'To', 'RecentDelays'])
df1['From'] = df1['From'].str.title()

## convert to first letter capital
df1['To'] = df1['To'].str.title()
df1
## split the delays
df2 = pd.DataFrame(df1['RecentDelays'].values.tolist(), columns=['Delay1','Delay2','Delay3'])

## remove column and merge 2 data frames--  final result
df1.drop('RecentDelays', axis=1, inplace=True)
pd.concat([df1, df2], axis=1)
```
